```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return 2 + 5 * x + 5 * np.random.randn(len(x))
```

```python
x = np.linspace(-3,3,20)
```

```python
p = f(x)
```

```python
plt.scatter(x,p);
```

```python
def loss_func(a,b):
    return np.sum((a + b * x - p)**2)/20
```

```python
loss_func(6,7)
```

```python
def gradient(a,b):
    return np.sum([2 * (a + b * x - p),2 * (a + b * x - p) * x],axis=1)/20
```

```python
gradient(6,7)
```

```python
a,b = 6,7
alpha = 0.1
N = 20
coeff = []
loss = []
for _ in range(N):
    delta = alpha * gradient(a,b)
    a = a - delta[0]
    b = b - delta[1]
    coeff.append((a,b))
    loss.append(loss_func(a,b))
    print(a,b)
    print(alpha * gradient(a,b))
    
```

```python
plt.plot(range(N),coeff);
```

```python
plt.plot(range(N),loss);
```
